## Ondas Mecnicas
### La ecuacin de onda en una dimensin espacial est dada por:
## $$ \dfrac{\partial^2 \phi}{\partial x^2} - \dfrac{1}{v^2}\dfrac{\partial^2 \phi}{\partial t^2}=0 $$
## con $\phi$ "la perturbacin, dada por $\phi=\phi(x,t)$ y $v$ la rapidez de la propagacin.


### La solucin a la EDP de onda est dada por:
### $$\phi(x,t)= A\cos(kw-wt) $$
### Con $A$, la amplitud, $k=\dfrac{2\pi}{\lambda}$, $w=\dfrac{2\pi}{T}$
### $\lambda \to $ longitud de onda
###  $T \to $ Periodo
