## Ondas Mecánicas
### La ecuación de onda en una dimensión espacial está dada por:
## $$ \dfrac{\partial^2 \phi}{\partial x^2} - \dfrac{1}{v^2}\dfrac{\partial^2 \phi}{\partial t^2}=0 $$
## con $\phi$ "la perturbación´´, dada por $\phi=\phi(x,t)$ y $v$ la rapidez de la propagación.


### La solución a la EDP de onda está dada por:
### $$\phi(x,t)= A\cos(kw-wt) $$
### Con $A$, la amplitud, $k=\dfrac{2\pi}{\lambda}$, $w=\dfrac{2\pi}{T}$
### $\lambda \to $ longitud de onda
###  $T \to $ Periodo
